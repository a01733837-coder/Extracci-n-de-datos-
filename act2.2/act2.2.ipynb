```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Convertir en dataframe
df= pd.read_csv("Actividad 1 Entendimiento de los Datos.csv")
df
```

```python
print(df.columns.tolist())
```

```python

 # a)Los registros de 5 “host name”  diferentes
cinco_hosts = df["property_type"].unique()
cinco_hosts
```

```python
#b) Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
host_2020 = df[df["host_since"]>"2020-12-31"]
host_2020


```

```python
#c) Los registros de los host que responden “a más tardar en 1 día”
host_1_dia = df[df["host_response_time"] == "within a day"]
host_1_dia

```

```python
#d) Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”
filas_mult_200 = df.iloc[199: :200]
filas_mult_200
```

```python
#e) Los registros de los tipos de cuarto: “Entire home”  y “Private room”
tipo_cuarto = df[df["room_type"].isin(["Entire home/apt", "Private room"])]
tipo_cuarto
```

```python
#f) Los registros que “a lo mucho cuenten con 4 camas” 
cuatro_camas = df[df["beds"]>=4]
cuatro_camas
```

```python
#g) Los registros que “sean superhost”
superhost = df[df["host_is_superhost"]=="t"]
superhost
```

```python
#h) Los registros de las columnas “impares”
columnas_impares = df.iloc[:, ::2]
columnas_impares
```
